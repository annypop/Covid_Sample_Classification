## ***Extracting Train data***

```python
xtrain = []
ytrain = []

classes = ['Cap', 'Cov', 'Norm']
rootPath = '/content/drive/My Drive/Colab Notebooks/DL Project/data'
print("Root: ", rootPath)
print("============================================================================================================")
print("Training Data : ")
for cls in classes:
  class_path = rootPath + '/Train/' + cls + '/'
  image_list = os.listdir(class_path)
  for image in image_list:
    img = cv2.imread(class_path + image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    xtrain.append(img)
    ytrain.append(cls)
```

## ***Extracting validation data***

```python
xvalid = []
yvalid = []

print("Validation Data : ")
for cls in classes:
  class_path = rootPath + '/Valid/' + cls + '/'
  image_list = os.listdir(class_path)
  for image in image_list:
    img = cv2.imread(class_path + image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    xvalid.append(img)
    yvalid.append(cls)
```

## ***Extracting Test data***

```python
xtest = []
ytest = []

print("Test Data : ")
for cls in classes:
  class_path = rootPath + '/Test/' + cls + '/'
  image_list = os.listdir(class_path)
  for image in image_list:
    img = cv2.imread(class_path + image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    xtest.append(img)
    ytest.append(cls)
```

```python
xtrain = np.array(xtrain).reshape()
ytrain = np.array(ytrain)
# ===========================
xvalid = np.array(xvalid)
yvalid = np.array(yvalid)
# ===========================
xtest = np.array(xtest)
ytest = np.array(ytest)
```
